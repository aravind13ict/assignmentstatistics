```python
#a. Find out how many males and females participated in the test

df.groupby('gender').size()
```

```python
#b. What do you think about the students' parental level of education?

df['parental level of education'].value_counts()
```

```python
# c Who scores the most on average for math, reading and writing based on gender


avgG = df[['gender', 'math score', 'reading score', 'writing score']].groupby('gender',).mean()
avgG
```

```python
# c. Who scores the most on average for math, reading and writing based on writing based on Test preparation course

avgTPC= df[['test preparation course', 'math score', 'reading score', 'writing score']].groupby('test preparation course').mean()
avgTPC
```

```python
# d the scoring variation for math,reading and writing based on gender

vargender= df[['gender', 'math score', 'reading score', 'writing score']].groupby('gender').std()
vargender
```

```python
# d the scoring variation for math,reading and writing based on test preparation course

varTPC= df[['test preparation course', 'math score', 'reading score', 'writing score']].groupby('test preparation course').std()
varTPC
```
